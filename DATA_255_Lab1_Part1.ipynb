## Wide and Deep Model

```python
# Splitting Deep continuous and categorical features
deep_categorical_features = ['Gender','Genres', 'Studios']  # Categorical features
deep_continuous_features = ['Age','Days Watched','Dropped', 'Rewatched', 'Popularity']  # Continuous features

# Getting column indices for wide, categorical, and continuous features from the DataFrame
wide_indices = [X_train.columns.get_loc(col) for col in wide_features]
deep_categorical_indices = [X_train.columns.get_loc(col) for col in deep_categorical_features]
deep_continuous_indices = [X_train.columns.get_loc(col) for col in deep_continuous_features]

wide_input = X_train_tensor[:, wide_indices]  # Extract wide features by column indices
deep_input_categorical = X_train_tensor[:, deep_categorical_indices]  # Extract categorical features by column indices
deep_input_continuous = X_train_tensor[:, deep_continuous_indices]  # Extract continuous features by column indices

print(f'Wide input shape: {wide_input.shape}')
print(f'Deep categorical input shape: {deep_input_categorical.shape}')
print(f'Deep continuous input shape: {deep_input_continuous.shape}')

wide_input_dim = X_train[wide_features].shape[1]  # 5 wide features
deep_input_categorical_dim = len(deep_categorical_features)
deep_input_continuous_dim = len(deep_continuous_features)


# Ensure the deep categorical input is correctly fed into embeddings
embedding_dims = [(len(anime_user_details[feature].unique()), min(50, (len(anime_user_details[feature].unique()) // 2))) for feature in deep_categorical_features]
print("Embedding Dims:", embedding_dims)
```

```python
class WideAndDeep(nn.Module):
    def __init__(self, wide_input_dim, embedding_dims, hidden_units, deep_input_continuous_dim, dropout_rate=0.3):
        super(WideAndDeep, self).__init__()

        # Wide layer (linear layer for memorization)
        self.wide = nn.Linear(wide_input_dim, 1)

        # Embedding layers for categorical features in the deep part
        self.embeddings = nn.ModuleList([nn.Embedding(input_dim, output_dim) for input_dim, output_dim in embedding_dims])

        # Calculate the correct deep input dimension
        total_embedding_dim = sum([output_dim for _, output_dim in embedding_dims])
        deep_input_dim_correct = total_embedding_dim + deep_input_continuous_dim

        # Deep part (MLP)
        self.deep = nn.Sequential(
            nn.Linear(deep_input_dim_correct, hidden_units[0]),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(hidden_units[0], hidden_units[1]),
            nn.ReLU(),
            nn.Dropout(dropout_rate),
            nn.Linear(hidden_units[1], 1)
        )

    def forward(self, wide_input, deep_input_categorical, deep_input_continuous):
        # Wide part forward pass
        wide_output = self.wide(wide_input)

        # Embed the categorical features
        embedded_categorical = [embedding(deep_input_categorical[:, i].long()) for i, embedding in enumerate(self.embeddings)]
        embedded_categorical = torch.cat(embedded_categorical, dim=1)

        # Concatenate embedded categorical features with continuous features
        deep_input_combined = torch.cat([embedded_categorical, deep_input_continuous], dim=1)

        # Pass through the deep network (MLP)
        deep_output = self.deep(deep_input_combined)

        # Combine wide and deep outputs
        output = wide_output + deep_output
        return output
```

```python
from torch.utils.data import DataLoader, TensorDataset
from tqdm import tqdm
import time

# Create a dataset with your input tensors
dataset = TensorDataset(wide_input, deep_input_categorical, deep_input_continuous, y_train_tensor)

# Define a DataLoader with batching
batch_size = 256 # You can adjust the batch size based on your system's memory
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

import torch
import os

# Check if CUDA is available, and if so, use the GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')  # This will print "cuda" if GPU is being used

# Initialize the model and move it to the GPU
model = WideAndDeep(
    wide_input_dim=wide_input_dim,
    deep_input_continuous_dim=deep_input_continuous_dim,  # Only continuous features
    embedding_dims=embedding_dims,  # Embeddings for categorical features
    hidden_units=[128, 64]
).to(device)  # Move model to GPU

# Initialize the criterion and optimizer
criterion = nn.MSELoss().to(device)  # Move loss function to GPU
optimizer = optim.Adam(model.parameters(), lr=0.0001, weight_decay=1e-4)

# Save checkpoint function
def save_checkpoint(model, optimizer, epoch, loss, filename='checkpoint.pth.tar'):
    checkpoint = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'loss': loss
    }
    torch.save(checkpoint, filename)

# Load checkpoint function (if needed for resuming training)
def load_checkpoint(model, optimizer, filename='checkpoint.pth.tar'):
    if os.path.isfile(filename):
        print(f"=> Loading checkpoint '{filename}'")
        checkpoint = torch.load(filename)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        epoch = checkpoint['epoch']
        loss = checkpoint['loss']
        print(f"=> Loaded checkpoint (epoch {epoch}, loss {loss})")
        return epoch  # Return the epoch to resume training
    else:
        print(f"=> No checkpoint found at '{filename}'")
        return 0

# Model training with tqdm for progress tracking and checkpointing
num_epochs = 20
checkpoint_dir = 'checkpoints/'  # Directory to save checkpoints

# Create the directory if it doesn't exist
if not os.path.exists(checkpoint_dir):
    os.makedirs(checkpoint_dir)

# Initialize tqdm progress bar
total_steps = len(dataloader) * num_epochs  # Total number of batches across all epochs

# Example: Model training with checkpointing and progress tracking using tqdm
with tqdm(total=total_steps, desc="Training Progress", unit="batch") as pbar:
    for epoch in range(num_epochs):
        model.train()
        epoch_loss = 0.0  # Track loss for each epoch

        # Loop over batches
        for batch_idx, (wide_batch, deep_cat_batch, deep_cont_batch, y_batch) in enumerate(dataloader):
            # Move inputs and targets to GPU
            wide_batch = wide_batch.to(device)
            deep_cat_batch = deep_cat_batch.to(device)
            deep_cont_batch = deep_cont_batch.to(device)
            y_batch = y_batch.to(device)

            optimizer.zero_grad()

            # Wide and Deep forward pass (model is already on GPU)
            predictions = model(wide_batch, deep_cat_batch, deep_cont_batch)

            # Loss calculation
            loss = criterion(predictions, y_batch)

            # Backward pass and optimization
            loss.backward()
            optimizer.step()

            # Accumulate loss for the current epoch
            epoch_loss += loss.item()

            # Update tqdm progress bar
            pbar.update(1)

            # Save a checkpoint every 100 batches (or adjust as needed)
            if batch_idx % 100 == 0:
                checkpoint_filename = os.path.join(checkpoint_dir, f'checkpoint_epoch_{epoch}_batch_{batch_idx}.pth.tar')
                save_checkpoint(model, optimizer, epoch, loss.item(), checkpoint_filename)

        # Save checkpoint at the end of each epoch
        checkpoint_filename = os.path.join(checkpoint_dir, f'checkpoint_epoch_{epoch}.pth.tar')
        save_checkpoint(model, optimizer, epoch, epoch_loss / len(dataloader), checkpoint_filename)

        # Print loss at the end of the epoch
        pbar.set_postfix({"Epoch": epoch+1, "Loss": epoch_loss / len(dataloader)})
        print(f'Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss / len(dataloader)}')

```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import torch

# Define the evaluation metrics function
def evaluate_metrics(y_true, y_pred):
    # Convert predictions to CPU if running on GPU
    if y_pred.is_cuda:
        y_pred = y_pred.cpu()
    if y_true.is_cuda:
        y_true = y_true.cpu()

    # Detach the tensors to convert them into numpy arrays
    y_pred = y_pred.detach().numpy()
    y_true = y_true.detach().numpy()

    # Calculate RMSE, MAE, R2 Score
    rmse = mean_squared_error(y_true, y_pred, squared=False)  # RMSE
    mae = mean_absolute_error(y_true, y_pred)  # MAE
    r2 = r2_score(y_true, y_pred)  # R2 Score

    return rmse, mae, r2
```

```python
import torch
import joblib

def recommend_anime(username, model, anime_user_details, X_scaler, top_n=10):
    # Filter the data to get all anime entries for the given username
    user_data = anime_user_details[anime_user_details['Username'] == username]

    if user_data.empty:
        print(f"No data found for username: {username}")
        return

    # Prepare wide, deep categorical, and deep continuous inputs
    wide_features = ['Username', 'anime_id', 'Genres', 'Studios', 'Status']
    deep_categorical_features = ['Gender', 'Genres', 'Studios']
    deep_continuous_features = ['Age', 'Days Watched', 'Dropped', 'Rewatched', 'Popularity']

    # Extract features for this user
    wide_input_data = user_data[wide_features]
    deep_input_categorical_data = user_data[deep_categorical_features]
    deep_input_continuous_data = user_data[deep_continuous_features]

    # Scale continuous features using the same scaler
    deep_input_continuous_data = X_scaler.transform(deep_input_continuous_data)

    # Convert to tensors
    wide_input_tensor = torch.tensor(wide_input_data.values, dtype=torch.float32)
    deep_input_categorical_tensor = torch.tensor(deep_input_categorical_data.values, dtype=torch.float32)
    deep_input_continuous_tensor = torch.tensor(deep_input_continuous_data, dtype=torch.float32)

    # Loading the saved model
    model = torch.load('wide_and_deep_model_final.pth')
    # Set the model to evaluation mode
    model.eval()

    # Disable gradient calculation for inference
    with torch.no_grad():
        # Move tensors to the appropriate device (GPU or CPU)
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        wide_input_tensor = wide_input_tensor.to(device)
        deep_input_categorical_tensor = deep_input_categorical_tensor.to(device)
        deep_input_continuous_tensor = deep_input_continuous_tensor.to(device)
        model = model.to(device)

        # Make predictions
        predictions = model(wide_input_tensor, deep_input_categorical_tensor, deep_input_continuous_tensor)

    # Inverse transform the predictions back to the original scale
    Inv_predictions = y_scaler.inverse_transform(predictions.cpu().numpy().reshape(-1, 1)).flatten()

    # Get the top N recommended anime based on prediction scores
    user_data['predicted_score'] = Inv_predictions
    recommended_anime = user_data.sort_values(by='predicted_score', ascending=False).head(top_n)

    # Return the recommended anime titles
    return recommended_anime[['Username','anime_id','English name']]

# Example usage:
X_scaler = joblib.load('X_scaler.pkl')
top_recommendations = recommend_anime(39745, model, anime_user_details, X_scaler, top_n=10)

if top_recommendations is not None:
    print("Top Anime Recommendations:")
    print(top_recommendations)
```
